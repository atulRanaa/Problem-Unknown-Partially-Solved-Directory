```python
arr = [0]*3321
for i in range(3321):
    s = str(df['TEXT'][i])
    arr[i] = len( [ c for c in s.split()] )
```

```python
max(arr)
```

```python
print(arr)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(figsize=(12,8))

axes.hist( arr, bins = 50, color='red', label ='Train (3321 samples)')
axes.legend()
axes.set_xlabel('Number of Words')
axes.set_ylabel('Samples Count')
axes.set_title('Word count distribution over Training Set');
```

```python
fig.savefig('hist.png', dpi = 100)
```

```python
text = [' ']*3321
for i in range(3321):
    s = str(df['TEXT'][i]).strip()
    l = len(s)
    if l > 0:
        text[i] = s[:20]
    else:
        text[i] = "null"
```

```python
print(text[:20])
```

```python
Set = list( set(text) )
print( len(Set))
count = [0] * len(Set)
```

```python
MAP = {}
for i in Set:
    MAP[i] = 0
```

```python
for i in text:
    MAP[i] += 1
dist_text_count = [0] * len(Set) 
for key in MAP:
    dist_text_count += [ MAP[key] ]
    
print(dist_text_count)
```

```python
common_text = []
for i in dist_text_count:
    if i >= 15:
        common_text += [ i]
print(len(common_text))
common_text.sort( reverse = True)
print(common_text)
```

```python
text_id = [ i for i in range( len(common_text) ) ]
fig, axes = plt.subplots(figsize=(12,8))

axes.bar(text_id,common_text, color='red')
axes.set_xlabel('Different Text IDs')
axes.set_ylabel('Samples Count')
axes.set_title('Common Text entry distribution over Training Set');
```

```python
fig.savefig('common_text.png', dpi = 100)
```
