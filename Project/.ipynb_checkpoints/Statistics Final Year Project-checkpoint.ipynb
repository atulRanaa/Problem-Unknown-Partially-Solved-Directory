```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('DataSet.csv')
```

```python
df.head()
```

```python
type(df['TEXT'][0])
```

```python
len(df)
```

```python
arr = [0]*3321
for i in range(3321):
    s = str(df['TEXT'][i])
    arr[i] = len( [ c for c in s.split()] )
```

```python
max(arr)
```

```python
print(arr)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(arr, bins = 50, color = 'grey')
plt.xlabel('Number of Words')
plt.ylabel('Samples Count')
plt.legend('Train (3321) samples')
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(12,8))

axes.hist( arr, bins = 50, color='grey', label ='Train (3321 samples)')
axes.legend()
axes.set_xlabel('Number of Words')
axes.set_ylabel('Samples Count')
axes.set_title('Word count distribution over Training Set');
```

```python
fig.savefig('hist.png', dpi = 100)
```

```python
text = [' ']*3321
for i in range(3321):
    s = str(df['TEXT'][i]).strip()
    l = len(s)
    if l > 0:
        text[i] = s[:10]
    else:
        text[i] = "null"
```
